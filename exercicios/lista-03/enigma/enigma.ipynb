# Exercício Enigma
**Enigma** é o nome de uma máquina usada para codificar mensagens. Veja mais informações [aqui](https://pt.wikipedia.org/wiki/Enigma_(m%C3%A1quina)). 

A máquina é útil na seguinte situação: uma pessoa deseja enviar uma mensagem para outra (no contexto de uma guerra, por exemplo, a mensagem poderia ser instruções sobre quando e onde atacar o inimigo) sem que uma terceira pessoa (o inimigo, no caso de uma guerra) possa ler o seu conteúdo. 

### Para transmitir uma mensagem - codificação
A mensagem é **codificada** com base em uma **Tabela de Códigos**. Em seguida a mensagem é enviada ao destinatário. Uma terceira pessoa pode interceptar a mensagem mas não vai compreender o seu significado pois, com a codificação, cada letra da mensagem original é substituída por outro símbolo. Então a mensagem fica sendo uma sequência aparentemente aleatória de letras.

A máquina Enigma era extremamente sofisticada mas, para este exercício, considere que a o processo de codificação e de codificação é bem simples e está baseado na ideia de que a Tabela de Códigos relaciona cada caracter a um outro caracter. Por exemplo, a letra *a* é codificada para o símbolo *T*, a letra *b* é codificada para o símbolo *+* , etc.

| Caracter | Código |
| -------- | ------ |
| a        |    T   |
| b        |    +   |



### Para ler uma mensagem - decodificação
O destinatário da mensagem deve **decodificar** a mensagem usando **a mesma Tabela de Códigos** que foi usada na codificação. Como resultado, obtém-se a mensagem original.


Para este exercício você deve implementar a classe `TabelaDeCodigos`  que representa uma **Tabela de Códgigos** e as funções `codifica` e `decodifica` que, respectivamente, codificam e decodificam uma mensagem, representada por uma *string*.

## Classe TabelaDeCodigos
A classe `TabelaDeCodigos` deve ter as seguintes características:

* ao ser criada pode-se:
    * informar um **dicionário** que relaciona os caracteres aos seus respectivos códigos. Neste caso uma tabela seria criada da seguinte forma: `tc = TabelaDeCodigos( {'a': 'T', 'b': '+' )`.
    * não informar nada. Neste caso a tabela deve usar um dicionário de códigos implementado diretamente no método construtor da classe. Esta tabela seria criada da seguinte forma: `tc = TabelaDeCodigos()`.
    
* possui dois atributos: um para armazenar o dicionário contendo os caracteres e seus respectivos códigos e outro para armazenar um dicionário no sentido contrário (dos códigos para os caracteres). Dicas:
    * `dic.items()`  retorna uma sequência de tuplas (chave, valor).
    * `dic.update({chave:valor})` adiciona uma nova entrada no dicionário.
    * `dic.get(chave)` retorna o valor associado à `chave` ou `None` para indicar que a chave não faz parte do dicionário.
    
* método que retorna o código associado a um caracter ou `None`  para indicar que o caracter não faz parte da tabela de códigos.

* método que retorna o caracter associada a um código ou `None` para indicar que o código não faz parte da tabela de códigos.

* método que valida a tabela de códigos. Uma tabela é considerada válida se para cada caracter houver um código único. Dica:
    * `len(dic)` retorna o número de chaves existente em um dicionário. Veja o que acontece com o método `update` quando se adiciona uma chave mais de uma vez.
    
    
## Função codifica
A função `codifica` deve ser usada para codificar mensagens.

### Parâmetros
Os parâmetros devem ser a mensagem (*string*) a ser codificada e uma tabela de códigos (objeto da classe `Tabela de Códigos`)

### Resultado
O resultado da função deve ser a mensagem codificada (na forma de *string*) ou None para indicar que não é possível codificar pois a mensagem possui pelo menos um caracter que não faz parte da tabela de códigos.


## Função decodifica
A função `decodifica` deve ser usada para decodificar mensagens.

### Parâmetros
Os parâmetros devem ser a mensagem (*string*) a ser decodificada e uma tabela de códigos (objeto da classe `Tabela de Códigos`)

### Resultado
O resultado da função deve ser a mensagem decodificada (na forma de *string*) ou None para indicar que não é possível decodificar pois a mensagem possui pelo menos um caracter que não faz parte da tabela de códigos.
