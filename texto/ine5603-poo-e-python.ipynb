# Programação Orientada a Objetos e Python
Relembrando, um programa de computador é um algoritmo escrito em alguma linguagem de programação. Um algoritmo, por sua vez, é uma sequência finita de ações que, quando realizadas, levam à solução de um problema. Para um mesmo problema podemos criar diversos algoritmos.

Na **Programação Orientada a Objetos (POO)** a ideia fundamental é que um problema pode ser resolvido por um "ser" chamado de **objeto**.

## Conceitos Fundamentais
Saber programar do jeito orientado a objeto depende de conhecer os seus conceitos fundamentais, que são: objeto, classe, atributo e método.

### Objeto
Um **objeto** é um ser capaz de:

* armazenar informações na sua memória.

* executar algoritmos.

### Classe
Todo objeto pertence à alguma classe. Fazendo uma analogia com a Biologia, o conceito de classe é equivalente ao conceito de **espécie**. Assim, objetos são os indivíduos de uma espécie.

Veja o desenho a seguir:

<img src="INE5603-objetos_classes.jpg">


O desenho mostra **6 objetos** (indivíduos) que pertencem à **3 classes** (espécies): pássaros, gatos e cachorros. Há três objetos da classe *gatos*, um objeto da classe *cachorro* e dois objetos da classe *pássaro*.

### Atributos
A palavra **atributo**, no contexto da POO, significa as informações que um objeto é capaz de armazenar na sua memória.

Veja que nós, humanos, somos dotados de memória, pois conseguimos lembrar de muitas coisas. Por exemplo, todos nós "sabemos de cor" o nosso nome, a cidade onde nascemos, quantos anos temos, etc. As informações *nome*, *cidade de nascimento* e *idade* são chamados de atributos. Em programação, usamos variáveis para armazenar as informações. Naturalmente, cada pessoa possui o seu nome, a sua cidade de nascimento, a sua idade, etc.

Os atributos válidos para um objeto são definidos na classe do objeto.

### Métodos
A palavra **método**, no contexto da POO, significa um algoritmo que um objeto é capaz de executar. 

Os métodos que um objeto pode executar são definidos na classe do objeto.


## Programar Orientado a Objetos
Programar orientado a objetos é, antes de tudo, **uma forma de pensar**. Esta forma é diferente da forma não orientada a objetos. Um programa de computador que seja formado apenas por funções, variáveis e comandos não é orientado a objetos. Para ser orientado a objetos o desenvolvedor tem que conscientemente raciocinar utilizando os quatro conceitos fundamentais (objeto, classe, atributo e método).

### Programação Orientada a Objetos e Python
A linguagem Python originalmente não era orientada a objetos. Somente a partir da versão 3 é que ela passou a ser orientada a objetos. Então, em termos práticos, é comum dizermos "use Python 2 se você não quer programar orientado a objetos e use Python 3 se quiser". 

Vamos compreender o pensamento orientado a objetos resolvendo um problema de duas maneiras: primeiro, sem usar os conceitos de POO e depois usando os conceitos de POO. Note que **qualquer problema** pode ser resolvido tanto do modo orientado a objetos como não orientado a objetos (que poderia ser, por exemplo, do modo conhecido como *programação estruturada*).


## Problema Exemplo: Terreno na praia
Considere o problema onde uma pessoa está pensando em comprar um terreno na praia e o corretor de imóveis oferece a ele dois terrenos. O corretor diz o código (por exemplo 'T035'), as dimensões e o preço de cada terreno. Para simplificar, considere que cada terreno tem o formato de um retângulo. Portanto as dimensões são a largura (chamada de frente) e o comprimento (chamado de fundos) do terreno.

Para ajudar a pessoa a decidir ela vai usar um programa de computador que a partir dos dados dos dois terrenos vai dizer qual é o mais barato. 

Note que o conceito de mais barato não depende apenas do preço do terreno. A área do terreno também é fundamental. Assim, mais barato significa o terreno que custar menos por metro quadrado.


### Solução 1 : Não Orientada a Objetos

O programa abaixo resolve o problema de um modo não orientado a objetos. O código abaixo deveria estar armazenado em um arquivo chamado, digamos, `programa_terreno_nao_oo.py` 

```python
# Programa que encontra o terreno mais barato
# Solução 1 : Versão não orientada a objeto

# função que calcula a área de um terreno
def calcula_area(comprimento, largura):
    return comprimento * largura

# função que calcula o valor do metro quadrado de um terreno
def calcula_preco_m2(preco, area):
    return preco / area

# função que determina qual preço por m2 é o menor. A função 
# retornará: o número zero para indicar que ambos preços
# por metro quadrado são iguais; o número -1 para indicar que o
# primeiro preço é menor que o segundo preço; o número +1 para
# indicar que o segundo preço é o menor o primeiro preço.
def determina_menor_preco_m2(preco_m2_1, preco_m2_2):
    if preco_m2_1 == preco_m2_2:
        resposta = 0
    else:
        if preco_m2_1 < preco_m2_2:
            resposta = -1
        else:
            resposta = 1
    return resposta

# Parte 1 : Definição dos dados problema

codigo_terreno1 = 'T035'
comprimento_terreno1 = 200   # 200 metros
largura_terreno1 = 50        # 50 metros
preco_terreno1 = 300000      # R$300.000,00

codigo_terreno2 = 'T104'
comprimento_terreno2 = 150   # 150 metros
largura_terreno2 = 80        # 80 metros
preco_terreno2 = 320000      # R$320.000,00


# Parte 2 : Resolução do problema
area_terreno1 = calcula_area(comprimento_terreno1, largura_terreno1)
preco_m2_terreno1 = calcula_preco_m2(preco_terreno1, area_terreno1)

area_terreno2 = calcula_area(comprimento_terreno2, largura_terreno2)
preco_m2_terreno2 = calcula_preco_m2(preco_terreno2, area_terreno2)

menor_preco_m2 = determina_menor_preco_m2(preco_m2_terreno1, preco_m2_terreno2)


# Parte 3 : Mostrar a solução do problema
print('Preço por m2 do terreno', codigo_terreno1, ': R$', preco_m2_terreno1)
print('Preço por m2 do terreno', codigo_terreno2, ': R$', preco_m2_terreno2)
if menor_preco_m2 == 0:
    msg = 'Os dois terrenos têm o mesmo preço por metro quadrado'
else:
    if menor_preco_m2 == -1:
        msg = 'O terreno ' + codigo_terreno1 + ' é o mais barato.'
    else:
        msg = 'O terreno ' + codigo_terreno2 + ' é o mais barato.'
print('Resposta:', msg)
    
```

### Solução 2 : Orientada a objetos
Na solução orientada a objetos os terrenos são vistos como objetos. Cada terreno é um objeto da "espécie" terreno. Assim o programador tem que pensar que existe uma classe que descreve as características de qualquer terreno.

Na solução abaixo é definida uma classe chamada `Terreno` que descreve as características dos terrenos: todo terreno é capaz de saber seu código, as suas dimensões e seu preço. Todo terreno consegue realizar ações pertinentes para um terreno: informar o seu código, calcular a sua área, calcular o valor do seu preço por metro quadrado. 

Observe que na *POO* todo objeto é considerado um ser vivo capaz de realizar ações (executar algoritmos) mesmo que isso não faça sentido na vida real. Terrenos, afinal, são coisas e não seres vivos.

O código abaixo deveria estar armazenado em um arquivo chamado, digamos, `programa_terreno_oo.py`.

```python
# Programa que encontra o terreno mais barato
# Solução 2 : Versão orientada a objeto

class Terreno:
    # armazena na memória do terreno o seu código, comprimento, largura e preco
    def __init__(self, codigo, comprimento, largura, preco):
        self.__codigo = codigo
        self.__comprimento = comprimento
        self.__largura = largura
        self.__preco = preco
        
    # obtém o código do terreno
    def obtenha_codigo(self):
        return self.__codigo
        
    # calcula a área do terreno    
    def calcule_area(self):
        return self.__comprimento * self.__largura
    
    # calcula o preço por metro quadrado do terreno
    def calcule_preco_m2(self):
        return self.__preco / self.calcule_area()
    
    # encontra o terreno com menor preço por m2.
    # O método retorna None para indicar que ambos os terrenos (self e outro_terreno) têm o 
    # mesmo preço por m2 ou retorna o terreno que tiver o menor preço por m2
    def encontre_mais_barato(self, outro_terreno):
        preco_m2_self = self.calcule_preco_m2()
        preco_m2_outro = outro_terreno.calcule_preco_m2()
        if preco_m2_self == preco_m2_outro:
            resposta = None  # ambos os terrenos têm o mesmo preço por m2
        else:
            if preco_m2_self < preco_m2_outro:
                resposta = self # o terreno self tem o menor preço por m2
            else:
                resposta = outro_terreno  # o terreno outro_terreno tem o menor preço por m2
        return resposta
        

# Parte 1 : Definição dos dados problema

terreno1 = Terreno('T035', 200, 50, 300000)
terreno2 = Terreno('T104', 150, 80, 320000)


# Parte 2 : Resolução do problema

terreno_mais_barato = terreno1.encontre_mais_barato(terreno2)


# Parte 3 : Mostrar a solução do problema
print('Preço por m2 do terreno', terreno1.obtenha_codigo(), ': R$', terreno1.calcule_preco_m2())
print('Preço por m2 do terreno', terreno2.obtenha_codigo(), ': R$', terreno2.calcule_preco_m2())
if terreno_mais_barato == None:
    msg = 'Os dois terrenos têm o mesmo preço por metro quadrado'
else:
    msg = 'O terreno ' + terreno_mais_barato.obtenha_codigo() + ' é o mais barato.'
print('Resposta:', msg)
```

#### Examinando a classe Terreno
A classe `Terreno` da solução 2, em termos de POO, possui:

* quatro atributos: 
    * `__codigo` 
    * `__comprimento`
    * `__largura`
    * `__preco`
* cinco métodos:
    * `__init__`
    * `obtenha_codigo`
    * `calcule_area`
    * `calcule_preco_m2`
    * `encontre_mais_barato`
    
##### Significado de self
A palavra `self` que aparece na definição da classe é uma variável que representa o objeto que está executando o método. Lembre que, na POO, é sempre um objeto quem executa um método. O nome da variável, `self`, é apenas uma convenção (todo mundo que programa em Python usa este nome).

##### Significado de __init__
O método `__init__` é também conhecido como **método construtor**. Seu objetivo, em Python, é: definir quais são os atributos da classe que está sendo definida e qual é o valor inicial destes atributos. Este método é executado automaticamente sempre que um objeto é criado. Assim, na linha `terreno1 = Terreno('T035', 200, 50, 300000)` está sendo criado um objeto da classe `Terreno` e o método `__init__` é executado.

O método `__init__` precisa ser definido em qualquer classe onde haja atributos.

##### Significado do método encontre_mais_barato
O método `encontre_mais_barato` implementa um algoritmo onde estão envolvidos dois objetos: o objeto `self`, que representa o objeto da classe `Terreno` que está executando o método e o objeto `outro_terreno`  que também é um objeto da classe `Terreno`.

É importante então que você compreenda que o algoritmo envolve dois objetos, dois terrenos. O objetivo do método é encontrar  qual dos dois terrenos, o `self` e o `outro_terreno`, é o mais barato. Literalmente, é o terreno `self` quem faz a análise e, no final, descobre se é ele ou se é o outro terreno quem tem o preço mais barato. Caso ambos os terrenos custem a mesma coisa então o método retorna o valor `None` para indicar que não é possível escolher um único terreno e então fica subentendido que os dois terrenos têm o mesmo preço por metro quadrado.

##### Localização da classe Terreno
Na solução mostrada a classe `Terreno` é definida no próprio arquivo `programa_terreno_oo.py`. Em uma situação real, no entanto, seria mais apropriado armazenar a classe em um arquivo chamado `terreno.py`. Com isso, da mesma forma como acontece com arquivos `.py`  que contêm definições de funções, a classe `Terreno` poderá ser utilizada em vários programas diferentes. Para isto bastará importar a definição: `from terreno import Terreno` e poder declarar variáveis como sendo do tipo `Terreno`.

#### Objetos da classe Terreno
Na solução orientada a objetos veja que são definidas duas variáveis, chamadas `terreno1` e `terreno2`, para representar os dois terrenos. São criados apenas dois objetos da classe `Terreno`  pois a expressão `Terreno(...)` aparece apenas duas vezes.

A variável `terreno_mais_barato` não representa um terceiro terreno pois isso não faria sentido (o problem envolve apenas dois terrenos). Assim, a variável é, digamos, apenas um sinônimo para um dos dois terrenos.

Observação avançada (apenas para programadores exerientes): as três variáveis (`terreno1`, `terreno2` e `terreno_mais_barato`) na verdade são ponteiros para objetos. Logo, `terreno_mais_barato` aponta para um dos dois terrenos ou não aponta para nenhum (quando seu valor é `None`).
