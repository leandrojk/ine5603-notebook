## Programa e Algoritmo
Um programa de computador é um texto cuja **forma** depende das regras gramaticas definidas para a linguagem de programação na qual o texto foi escrito. Mas qual deve ser o conteúdo deste texto?

**Um programa de computador é um algoritmo escrito em alguma linguagem de programação**. É preciso, portanto, compreender o que é um **algoritmo**

## Algoritmo
Um algoritmo **é um modo de resolver problemas**, um modo de pensar chamado de **pensamento algorítmico**. Este modo de pensar existe "desde sempre" e está relacionado como os humanos resolvem problemas matemáticos.

A palavra **algoritmo** surgiu no século IX com o matemático islâmico **Abu Ja’far Muhammad ibn
Musa al-Khwarizmi** no Iraque.

![Abu](https://www.researchgate.net/profile/Jorge_Fernandez25/publication/326357560/figure/fig13/AS:647700987011079@1531435357496/Abu-Jafar-Muhammad-ibn-Musa-al-Khwarizmi.png)

No século IX, Abu escreveu um livro (em árabe) mostrando o sistema indiano de representação numérica (sistema decimal).

No século XII o livro foi traduzido para o latim e o sistema indo-arábico começou a ser usado na Europa. O sistema romano de numeração (os famosos números romanos que aprendemos na escola) foi sendo abandonado.

## Algoritmos e Matemática
Desde de sempre, mas especialmente desde o século IX, a idéia de algoritmo esteve ligada a resolução de problemas matemáticos.

Um problema pode ser resolvido por meio algorítmico se for possível definir uma sequência finita de ações que levam à solução do problema. Esta sequência deve funcionar para qualquer caso.

Oposto de solução algorítmica: resolver o problema de forma improvisada, com base na tentativa e erro, chegando na solução por acaso, se der sorte.

Algoritmos: um modo de pensar! É preciso aprender a pensar deste modo!!

Qualquer programa de computador, sobre qualquer assunto, é um algoritmo escrito em uma linguagem de programação.

Problemas matemáticos simples são perfeitos para se aprender a desenvolver o pensamento (raciocínio) algorítmico.

Para um mesmo problema sempre há diversos algorítmos possíveis. A graça da profissão de quem faz Computação
ou Sistemas de Informação está em tentar encontrar o algoritmo mais simples, que produz a a solução do problema com o menor esforço. Sempre queremos que o problema seja resolvido no menor tempo possível. É mais ecológico (o
computador consumirá menos energia elétrica)!

No mundo ideal, algoritmos são perfeitos: resolvem qualquer problema sem erros, sem produzir resposta errada. No mundo real, porém, é extremamente difícil criar algoritmos perfeitos:

* Escreve-se o algoritmo.
* Escreve-se o programa (algoritmo escrito em alguma linguagem de programação).
* Assume-se que ele é perfeito. Até que um dia um usuário reclama que o programa não funcionou. Surge o bug (o programa "tá bugado").
* Corrige-se o bug alterando o algoritmo.
* Sai a nova versão do programa, torcendo para que ele não contenha erros.




## Exercício: Somar dois números
Considere como primeiro exercício a tarefa de resolver o problema matemático *calcular a soma de dois números*. 

Você, como universitário, deve considerar este problema muito simples. Mas lembre-se que um dia, há muitos anos, este era um problema díficil nas aulas de matemática no ensino fundamental. Por exemplo, quanto dá 537 + 342? Note que a resposta não sai imediatamente, você tem que pensar um pouco. Este ato de pensar significa que você, na sua cabeça, está executando um algoritmo, um procedimento, para conseguir chegar na resposta.

### Algoritmo 1:
Um algoritmo para resolver o problema de somar dois números poderia ser o seguinte:

1. Escreva a expressão numérica que representa a soma desejada.
1. Expanda a expressão numérica substituindo cada número por outra expressão numérica. Estas duas novas expressões devem representar os respectivos números decompostos em centenas, dezenas e unidades.
1. Reorganize a expressão numérica obtida na etapa anterior fazendo que as centenas fiquem ao lado uma da outra. Faça o mesmo com as dezenas e com as unidades.
1. Some as centenas, as dezenas e as unidades produzindo uma nova expressão numérica.
1. O resultado final da soma será formada pelos primeiros dígitos da centena, dezena e unidade obtidos na etapa anterior.

Um algoritmo nada mais é do que um conjunto de instruções, neste exemplo escritas na forma de frases em português. Se a pessoa seguir direitinho as instruções poderá resolver o problema de somar dois números.

Por exemplo, usando o algoritmo para descobrir quanto dá a soma 537 + 342 deveríamos proceder da seguinte maneira:

````
537 + 342  (passo 1)

500 + 30 + 7  +  300 + 40 + 2  (passo 2)

500 + 300 + 30 + 40 + 7 + 2   (passo 3)

800 + 70 + 9    (passo 4)

879   (passo 5)

A solução do problema é o número 879.

````

Observe que a estratégia empregada no algoritmo consiste em trocar uma conta difícil (537 + 342) por várias outras contas fáceis (500 + 300, 30 + 40  e 7 + 2). 

Agora vamos usar o mesmo algoritmo para resolver outro problema: 382 + 246. 

````
382 + 246   (passo 1)

300 + 80 + 2 + 200 + 40 + 6  (passo 2)

300 + 200 + 80 + 40 + 2 + 6  (passo 3)

500 + 120 + 8   (passo 4)

518  (passo 5)

A solução do poroblema é o número 518.
````

Observe que para este exemplo o algoritmo não funcionou, mesmo a pessoa tendo seguido exatamente o que determinava o algoritmo. Quem está executando um algoritmo não deve interpretar/alterar as instruções (os passos). Temos então a situação onde o algoritmo funciona para alguns casos e não funciona para outros. Portanto, o algoritmo não é válido pois para ser chamado de algoritmo ele deveria funcionar para somar quaisquer dois números.

Precisamos então corrigir o algoritmo. Para isso precisamos identificar qual passo está errado. No exemplo, o problema está no passo 4. O erro acontece sempre que a soma das centenas for maior que 100, ou a soma das dezenas for maior que 90, ou a soma das unidades for maior que 10.

Podemos modificar o passo 4 de modo que sempre que um dos casos acima acontecer é preciso voltar a executar a partir do passo 2.

### Algoritmo 1a:
A versão corrigida do algoritmo 1 poderia fica assim:


1. Escreva a expressão numérica que representa a soma desejada.
1. Expanda a expressão numérica substituindo cada número por outra expressão numérica. Estas duas novas expressões devem representar os respectivos números decompostos em centenas, dezenas e unidades.
1. Reorganize a expressão numérica obtida na etapa anterior fazendo que as centenas fiquem ao lado uma da outra. Faça o mesmo com as dezenas e com as unidades.
1. Some as centenas, as dezenas e as unidades produzindo uma nova expressão numérica.
1. Se a soma das centenas tiver mais de três dígitos ou a soma das dezenas tiver mais de dois dígitos ou a soma das unidades tiver mais de um dígito então continue a executar o algoritmo a partir do passo 2.
1. O resultado final da soma será formada pelos primeiros dígitos da centena, dezena e unidade obtidos na etapa anterior.

Vamos testar o novo algoritmo somando 382 e 246:

````

500 + 100 + 20 + 8   (passo 2)

600 + 20 + 8   (passo 4)

analisando os números que todos estão ok  (passo 5)

628   (passo 6)

A solução do problema é 628.
````
### O algoritmo está sempre correto?
O novo algoritmo agora funcionou para o caso que havia falhado. Mas podemos garantir que vai funcionar para a soma de dois números quaisquer? O que as empresas fazem é testar o algoritmo para muitos casos e, no final, dizer algo do tipo "testamos bastante e, até onde vimos, não encontramos erros. Se você encontrar um erro nos avise que iremos providenciar a correção".

Nenhuma empresa pode afirmar que o seu software é 100% correto pois, se afirmasse, poderia ser processada por uma pessoa que encontrasse uma falha. Na verdade, o que muitas empresas fazem é exatamente premiar (pagando em dinheiro) pessoas que encontrarem falhas.

No nosso caso, o algoritmo só funciona se a soma der no máximo 999.

## Algoritmos funcionam porque...
Quem executa um algoritmo, uma pessoa ou uma máquina, **não precisa compreender por que o algoritmo consegue resolver o problema**. Basta que a pessoa ou máquina saiba executar cegamente a sequência de passos.

Com algoritmos, resolve-se um problema difícil pela aplicação sistemática de um conjunto de ações  bem mais simples (fáceis de serem executadas).

## Exercício: Encontrar os divisores de um número inteiro positivo
Considere agora o problema de encontrar os divisores de um número inteiro positivo. Será que existe algum procedimento sistemático que consiga encontrar os divisores?

Antes de pensarmos em um algoritmo é importante compreender exatamente o problema que queremos resolver. Para isso o conceito matemático de divisor de um número deve estar claro: um número é divisor de outro número se o resto da divisão inteira entre o outro número e o número for igual a zero. Por exemplo, 4 é divisor de 12 porque dividindo 12 por 4 dá 3 e o resto é zero. Já 7 não é divisor de 12 porque dividindo 12 por 7 dá 1 e o resto é 5.

Qual estratégia podemos usar para encontrar os divisores de um número N? Uma ideia seria testar cada um dos números existentes entre 1 e N. O teste consiste em verificar se o resto da divisão de N pelo número é zero. Se for então o número é divisor de N.

### Algoritmo 3
Para encontrar os divisores de um número N faça o seguinte:

1. Para cada um dos números existentes entre 1 e N:
   1. Calcule o resto da divisão de N pelo número
   1. Se o resto da divisão for zero então o número é um divisor de N
   
Por exemplo, para encontrar os divisores de 6 usando o algoritmo:

````
O primeiro número a ser testado é o 1    (passo 1)

6 / 1 = 6, resto zero     (passo 1.1)

Como o resto é zero então o número 1 é divisor de 6    (passo 1.2)



O segundo número a ser testado é o 2    (passo 1)

6 / 2 = 3, resto zero     (passo 1.1)

Como o resto é zero então o número 2 é divisor de 6    (passo 1.2)



O terceiro número a ser testado é o 3    (passo 1)

6 / 3 = 2, resto zero     (passo 1.1)

Como o resto é zero então o número 3 é divisor de 6    (passo 1.2)



O quarto número a ser testado é o 4    (passo 1)

6 / 4 = 1, resto 2     (passo 1.1)

Como o resto é 2 o número 4 não é divisor de 6    (passo 1.2)



O quinto número a ser testado é o 5    (passo 1)

6 / 5 = 1, resto 1     (passo 1.1)

Como o resto é 1 o número 5 não é divisor de 6    (passo 1.2)



O sexto número a ser testado é o 6    (passo 1)

6 / 6 = 1, resto zero     (passo 1.1)

Como o resto é zero o número 6 é divisor de 6    (passo 1.2)

````

O algoritmo 3 funciona e encontra os divisores de qualquer número. Mas, pensando bem, ele pode ser melhorado em dois aspectos:

* Os números 1 e N sempre serão divisores de N. Logo, não é necessário fazer nenhum cálculo para concluir que são divisores.

* Nenhum número maior que N / 2 será divisor de N. Logo, não é necessário testar os números que sejam maiores que N / 2.

Cosiderando as observações acima, podemos criar um novo algoritmo, inspirando pelo Algoritmo 3, que encontra os divisores de N fazendo pouco menos da metade dos cálculos realizados no Algoritmo 3.

### Algoritmo 4:
Para calcular os divisores de um número N faça o seguinte:

1. 1 é divisor de N.
1. Para cada um dos números existentes entre 2 e N / 2 faça:
   1.1 Calcule o resto da divisão entre N e o número.
   1.2 Se o resto da divisão for igual a zero então o número é um divisor de N.
1. N é divisor de N.

Algoritmos muitas vezes são comparados à receitas culinárias: a partir dos ingredientes (dados do problema a ser resolvido), ações são realizadas de modo que, no final, temos o prato pronto para ser servido (a solução do problema).

Pessoas seguindo uma receita utilizam o **bom senso** quando ficam em dúvida. **Computadores não têm bom senso!** Tudo precisa ser bem detalhado. Computadores não compreendem as intenções de quem escreveu o algoritmo, eles executam cegamente as instruções.

### Definição de algoritmo
Um algoritmo é uma **sequência finita de ações** que, quando executadas, resolvem um problema.

#### Sequência
As ações são executadas uma após a outra. Exsite, portanto, a noção de primeira ação, segunda ação, etc.

#### Finita
O número de ações a ser executado sempre é finito. Pode ser, dependo do problema, uma quantidade muito grande de ações, que demorarão horas, dias, meses, anos, para serem executadas. Mas, um dia chegam a um fim.

#### Ações
O computador irá fazer exatamente o que lhe for mandado.

Os conceitos de *sequência* e de *finitude* são fáceis de entender. Mas o que é uma ação? Quais são as ações possíveis? O que exatamente um computador sabe fazer sozinho?

## Comandos
As ações que um computador consegue executar são chamadas de **comandos**. Assim, literalemente, o programador comanda a máquina determinando que ela **execute** uma sequência de comandos.
